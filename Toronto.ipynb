### Segmenting and clustering neighborhoods in Toronto

First let's scrape the information using pandas

```python
import pandas as pd

url= 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
tables= pd.read_html(url)
```

```python
# show the head and take out the Boroughs not assigned
tables[0]
df = tables[0][tables[0]['Borough'] != 'Not assigned']
df.head()
```

```python
# More than one neighborhood can exist in one postal code area, combined these into one row with the neighborhoods separated with a comma
df["Neighbourhood"] = df.groupby("Postcode")["Neighbourhood"].transform(lambda neigh: ', '.join(neigh))

#remove duplicates
df = df.drop_duplicates()

    
df.head()
```

```python
# If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough
df['Neighbourhood'].replace("Not assigned", df["Borough"],inplace=True)
df
```
